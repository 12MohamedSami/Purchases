# Merge the Customer table with the Orders table

```python
pd.read_sql('''SELECT customer_city , customer_state , order_purchase_timestamp, order_approved_at ,order_estimated_delivery_date
               FROM customers C JOIN orders O
               ON C.customer_id= O.customer_id LIMIT 20;''',data)
```

# **Total Price for each Product**

```python
pd.read_sql('''SELECT  product_category_name Product_Name, SUM(price) Total_Price , product_description_lenght Description_length , shipping_limit_date
               FROM order_items O_I JOIN products P
               ON O_I.product_id = P.product_id
               WHERE product_category_name NOT LIKE '%None%' AND product_description_lenght NOT LIKE '%NaN%'
               GROUP BY product_category_name
               ORDER BY price DESC;''',data)
```

```python
pd.read_sql('''SELECT DISTINCT business_type, declared_monthly_revenue
               FROM leads_closed WHERE declared_monthly_revenue NOT in (0);''',data)
```

```python
pd.read_sql('''SELECT DISTINCT business_type, AVG(declared_monthly_revenue) Average_Monthly_Revenue
               FROM leads_closed WHERE declared_monthly_revenue NOT in (0)
               GROUP by business_type;''',data)
```

```python
pd.read_sql('''SELECT first_contact_date First_Date , won_date Won_Date , origin  , business_segment,  business_type, lead_type, lead_behaviour_profile
               FROM leads_closed LC JOIN leads_qualified LQ
               ON LC.mql_id= LQ.mql_id
               WHERE lead_behaviour_profile NOT NULL;''',data)
```

```python
pd.read_sql('''SELECT order_status, review_comment_title, review_comment_message, review_score
               FROM order_reviews O_R JOIN orders O
               ON O_R.order_id = O.order_id
               WHERE review_comment_title AND review_comment_message NOT NULL;''',data)
```
